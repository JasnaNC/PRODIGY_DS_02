### Imputing the missing values in the dataset

```python
from scipy import stats
data['Age'].fillna(data['Age'].median(),inplace=True)
mode=stats.mode(data['Embarked'].dropna())[0][0]
data['Embarked'].fillna(mode,inplace=True)
mode_cab=stats.mode(data['Cabin'].dropna())[0][0]
data['Cabin'].fillna(mode_cab,inplace=True)
```

## Correlation analysis
